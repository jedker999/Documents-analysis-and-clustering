# PREPROCESSING

```python
ps = PorterStemmer()
counter = 0
for filename in glob.glob('dane/*.json'):
    with codecs.open(filename, 'r', 'utf-8') as f: 
        article = str(json.load(f)) 
        #removing stop-words and useless characters
        stop_words = set(stopwords.words('english'))
        words = word_tokenize(article)
        filtered_sentence = []
        digits = str(list(range(1,2100)))
        punct=str(string.punctuation)
        letters = list(string.ascii_uppercase)
        lletters = list(string.ascii_lowercase)
        
        for w in words:
            if w not in stop_words and w not in digits and w not in letters and w not in lletters and w not in punct:
                filtered_sentence.append(w)

        #stemming
        stem_sentence = []
        for w in filtered_sentence:
            stem = ps.stem(w)
            stem_sentence.append(stem)

        #useless words, strange characters
        useless_words =["'", '``', '••', '≤', '≥', 'ϫ', '~', "'•", "•", "'••", "'", "''", "'/", "'abstract", 'abstract', 'none', "'none", "'section", "'ref_span", "'cite_span", "td", "/tr", "/td", "the", "'volum", "--", "'other_id", "'year", 'Ϯ', "'author", "'titl", "'start", "'end", "'page", "'text", "'ref_id", "'last", "'first", 'author', 'text', 'suffix', "'suffix", "'middl", "\\xa0"]
        def useless_w(stem_sentence, useless_words):
            return [w for w in stem_sentence if w not in useless_words]

        wlist = useless_w(stem_sentence, useless_words)
     
        lista = []
        for ele in wlist:
            x = re.sub("^[']*|^[-]*|[-]*$|^[/]*|[/]*$|^[~]*|[~]*$|^[\.]*|[\.]*$", "", ele)  
            y = re.sub("^bibref.*$|^figref.*$|^tabref.*$|^\w{1,2}$|^[+-a-zA-Z0-9]*([0-9]*[.])?[0-9]+$|^:.*$|^[0-9]{1,10}$|^[a-zA-Z][0-9]{1,4}$|^[0-9][a-zA-z]$|^[a-zA-Z]?.?$|^[0-9]*[:-][0-9]*[:-][0-9]*.*$|^[0-9]*-[0-9]*[\w]*$|^[a-zA-Z0-9]*[\.:-][a-zA-Z0-9]*[\.:-][a-zA-Z0-9]*[a-zA-Z0-9]*$|^[a-zA-Z0-9]*[\.:-][a-zA-Z0-9]*[\.:-][a-zA-Z0-9]*[a-zA-Z0-9]*[\.:-][a-zA-Z0-9]*|^[+]*[a-z]?$|[+]*$|^[0-9]*\.[a-z][0-9]$|^[0-9]{1,5}[a-z]$|^[a-z]{1,2}[0-9]{1,5}$|^venu$|^issn$|^der$", "", x)
            if y:               
                lista.append(y) 

        
        filename = 'preprocessing/plik'+str(counter)+'.txt'
        os.makedirs(os.path.dirname(filename), exist_ok=True)
        with open(filename,"w", encoding='utf-8') as plik:
            plik.write(str(lista))    
        plik.close()
        counter = counter + 1   
print(lista)
```

```python
#removing topwords and preprocessing 2
from collections import Counter
from glob import iglob
import glob
import re
import os
import math


def remove_garbage(text):
    text = re.sub(r'\W+', ' ', text)
    text = text.lower()
    return text

filepath2 = 0
topwords = 2000
a = []
counter = Counter()
for filepath in iglob('preprocessing/plik*.txt'):
    with open(filepath, encoding='utf-8') as file:
        #print(filepath) 
        filepath2 = filepath2 + 1
        counter.update(remove_garbage(file.read()).split())

        for word, count in counter.most_common(topwords):
            #print('{}: {}'.format(count, word)
            a.append(word)
#print(len(a))
  
coun = Counter(a)
rw=[]
for word, count in coun.most_common(len(a)):
    #print('{}: {}'.format(count, word))                      
    proc = 70
    v = (filepath2*proc)/100
    v = math.ceil(v)
    #print(v)     
    if v<=count:
        rw.append(str(word))

        
counter = Counter()      
counter2 = 0
for filename in glob.glob('preprocessing/plik*.txt'):
    with codecs.open(filename, 'r', 'utf-8') as f: 
        content = f.read().split()
        final = []  
        a=[]      
        for word in content:
            y = re.sub("\"|,|'", "", word)
            if y:                 
                a.append(y)  
        f.close()                
        for ele in a:
            if ele not in rw:
                final.append(ele)       
        
        filename = 'preprocessing2/plik'+str(counter2)+'.txt'
        os.makedirs(os.path.dirname(filename), exist_ok=True)
        with open(filename,"w", encoding='utf-8') as plik:
            plik.write(str(final))    
        plik.close()
        counter2 = counter2 + 1  
        print(final)
```

# DOC2VEC

```python
from glob import glob
from os.path import basename, splitext

import pandas as pd
from gensim.corpora import Dictionary
from gensim.models import LsiModel
from gensim.similarities import MatrixSimilarity
import math


#ref material https://tedboy.github.io/nlps/gensim_tutorial/tut3.html

texts = {}  
cou = 0
for filename in glob("preprocessing2/*.txt"):
    with open(filename, "r", encoding="utf-8") as f:
        cou = cou +1
      
        doc = (
            " ".join(f.readlines())
            .replace("[", "")
            .replace("]", "")
            .replace(",", "")
            .replace("'", "")
            .split()
        )

        
        texts[splitext(basename(filename))[0]] = doc



dictionary = Dictionary(texts.values())
corpus = [dictionary.doc2bow(text) for text in texts.values()]
lsi = LsiModel(corpus, id2word=dictionary)
index = MatrixSimilarity(lsi[corpus])

# similarity
similarity_array = []
for filename, doc in texts.items():
    vec_bow = dictionary.doc2bow(doc)
    vec_lsi = lsi[vec_bow]
    sims = index[vec_lsi]
    similarity_array.append(sims)


similarity_df = pd.DataFrame(
    data=similarity_array, index=texts.keys(), columns=texts.keys()
)
#similarity_df.stack().reset_index(inplace=True)

#print(similarity_df)
proc = 25
w = (cou*proc)/100
w = math.ceil(w)
new_df = similarity_df.stack().reset_index(level=1, name='wynik').rename(columns={'level_1':'plik1'})[['wynik','plik1']]
new_df['plik2'] = new_df.index
new_df.sort_values(by=['wynik'], inplace=True, ascending=False)
data = new_df.loc[new_df['plik1'] != new_df['plik2']]
data = data.iloc[::2]
data = data.reset_index(drop=True)
data.plik1 = data.plik1.str.replace(r'plik', r'')
data.plik2 = data.plik2.str.replace(r'plik', r'')
ds = data.head(w)
#print(ds)


ds2 = ds[ds.wynik < 0.9]
print(ds2)
```

# TF-IDF analysis

```python
for index, row in ds2.iterrows():
    plik1 = row['plik1']
    plik2 = row['plik2']
    docs = ['preprocessing2/plik'+str(plik1)+'.txt', 'preprocessing2/plik'+str(plik2)+'.txt']

    def extract(filename):
        with open(filename,encoding="utf8") as f:
            features = []
            for line in f:
                features += line.strip().split(', ')
            return features

    vectorizer = TfidfVectorizer(analyzer=extract).fit(docs)
    vectors = vectorizer.fit_transform(docs)
    feature_names =vectorizer.get_feature_names()
    dense = vectors.todense()
    denselist = dense.tolist()
    words = word_tokenize(str(denselist)) 
    useless_words =["'",",", "''", "["]
    def useless_w(stem_sentence, useless_words):
        return [w for w in stem_sentence if w not in useless_words]

    words2 = useless_w(words, useless_words)    
    df = pd.DataFrame(denselist, columns=feature_names)
    pd.set_option("display.max_rows", None, "display.max_columns", None)     
    df2 = df[df > 0]
    df3 = df2[df2.columns[~df2.isnull().any()]]
    df4 = df3.mean()
    df5 = df4[df4 > 0.07]
    print("plik"+str(plik1)+" x "+"plik"+str(plik2))
    print(df5.sort_values(ascending = False))
    print()    

    filename = 'tfidf/tfidf'+str(plik1)+','+str(plik2)+'.txt'
    os.makedirs(os.path.dirname(filename), exist_ok=True)
    with open(filename,"w", encoding='utf-8') as plik:
        plik.write(str(df))    
    plik.close()
```
